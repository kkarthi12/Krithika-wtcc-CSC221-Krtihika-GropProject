```python
import pandas as pd

#Read Dataframe from CSV file 'CSC221-webscrape-data.csv,
pop_df = pd.read_csv('CSC221-GP-data.csv')
print(pop_df.shape)

# DROP LAST ROW - DON'T NEED 'all' Values 
pop_df = pop_df.drop(pop_df.index[-1])

print(pop_df.columns)

# Display all the data in the file 
pop_df.head(12) 
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Remove extra spaces from column names
pop_df.columns = pop_df.columns.str.strip()
# Remove rows with missing values
pop_df = pop_df.dropna(subset=['Year', 'Top 0.001%'])

# Check the column names
print(pop_df.columns)

#pop_df['Top'] = pop_df['Top'].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip()

pop_df['Top 0.001%'] = pop_df['Top 0.001%'].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip()
# Convert data to numeric
pop_df['Year'] = pd.to_numeric(pop_df['Year'], errors='coerce')
pop_df['Top'] = pd.to_numeric(pop_df['Top 0.001%'], errors='coerce')

# Create the bar graph
plt.figure(figsize=(8, 5))
plt.bar(pop_df['Year'], pop_df['Top'])
 

plt.title('Top by Year')
plt.xlabel('Year')
plt.ylabel('Top')
plt.xticks(rotation=45)

plt.show()

```
